```python
class EulerPredictor:

    def __init__(
        self,
        pred_step_size: float,
        prediction_horizon: int,
    ):
        self.pred_step_size = pred_step_size
        self.dim_input = 2
        self.prediction_horizon = prediction_horizon
        

    def predict(self, current_state, action, impacted_price):
        next_state = np.zeros(2)
        next_state[0] = current_state[0] - self.pred_step_size*action*3
        next_state[1] = current_state[1] + self.pred_step_size*action*(impacted_price - current_state[1])
        return next_state

    def predict_sequence(self, state, action_sequence, impacted_price):
        observation_sequence_predicted = np.zeros([self.dim_input, self.prediction_horizon])
        current_state = state

        for k in range(self.prediction_horizon):
            current_action = action_sequence[:, k]
            next_state = self.predict(current_state, current_action, impacted_price)
            observation_sequence_predicted[:, k] = next_state
            current_state = next_state
        
        return observation_sequence_predicted
```

```python
def running_objective(state, action, impacted_price):
    return (((state[0] -48)/state[0])**2 + ((state[1]-impacted_price)/state[1])**2)
```

```python
class ControllerMPC:
    def __init__(self, candles, i, predictor, optimizer, action_bounds=None, prediction_horizon = 5, discount_factor = 0.5):
        super().__init__()
        self.predictor = predictor
        self.dim_output = 1
        self.prediction_horizon = prediction_horizon
        self.discount_factor=discount_factor
        self.optimizer = optimizer
        self.action = np.zeros(self.dim_output)
        self.action_sequence = np.squeeze(np.tile(self.action, (1, self.prediction_horizon + 1)))
        self.action_bounds = np.tile(action_bounds.T, (self.prediction_horizon + 1))
        self.running_objective = running_objective
        volume = candles.iloc[i].volume
        self.impacted_price = candles.iloc[i].impacted_price
        price = candles.iloc[i].mid_price
        self.observation = np.array([volume, price])

    def mpc_objective(
        self,
        action_sequence, 
        observation
    ):

        action_sequence_reshaped = np.reshape(action_sequence, [self.prediction_horizon + 1, self.dim_output]).T

        observation_sequence_predicted = self.predictor.predict_sequence(observation, action_sequence_reshaped, self.impacted_price)

        observation_sequence = np.column_stack((observation, observation_sequence_predicted))

        actor_objective = 0

        for k in range(self.prediction_horizon + 1):
            actor_objective += self.discount_factor**k * self.running_objective(
                observation_sequence[:, k], action_sequence_reshaped[:, k], self.impacted_price
            )
        return actor_objective

    def compute_action(
        self
    ):
        self.action_sequence = self.optimizer.optimize(lambda action: self.mpc_objective(action, self.observation), self.action_sequence, self.action_bounds)
        return self.action_sequence[:1]
```

```python
from rcognita.optimizers import SciPyOptimizer

predictor = EulerPredictor(
    pred_step_size=1,
    prediction_horizon=5
)

optimizer = SciPyOptimizer(
    opt_options={
        "maxiter": 100,
        "maxfev": 5000,
        "disp": False,
        "adaptive": True,
        "xatol": 1.0E-7,
        "fatol": 1.0E-7,
    },
    opt_method="SLSQP"
)
```

```python
def execute_buy_control(candles, cash, i, to_buy):
    asks = candles.loc[i, 'asks']
    asks = np.array(ast.literal_eval(asks))
    j = 0
    bought = 0
    spent = 0
    

    controller = ControllerMPC(
        candles,
        i,
        predictor = predictor, 
        optimizer = optimizer, 
        action_bounds=np.array([[0, 1]]), 
        prediction_horizon = 5,
        discount_factor = 0.5
    )
    
    action = controller.compute_action()
    while j < asks.shape[0] - 1 and bought + float(asks[j, 1]) < to_buy * action:
        bought += float(asks[j, 1])
        spent += float(asks[j, 0]) * float(asks[j, 1])
        j += 1


    candles.loc[i, 'impacted_price'] = float(asks[j, 0])



    cash -= spent 
    candles.loc[i, 'position'] += bought
    candles.loc[i, 'action'] = action

    return candles, cash, action

def execute_sell_control(candles, cash, i, to_sell):
    bids = candles.loc[i, 'bids']
    bids = np.array(ast.literal_eval(bids))  
    j = 0
    sold = 0
    gained = 0

    controller = ControllerMPC(
        candles,
        i,
        predictor = predictor, 
        optimizer = optimizer, 
        action_bounds=np.array([[0, 1]]), 
        prediction_horizon = 5,
        discount_factor = 0.5
    )

    action = controller.compute_action()
    while j < bids.shape[0] - 1 and sold + float(bids[j, 1]) < to_sell * action:
        sold += float(bids[j, 1])
        gained += float(bids[j, 0]) * float(bids[j, 1])
        j += 1


    candles.loc[i, 'impacted_price'] = float(bids[j, 0])

    
    cash += gained 
    candles.loc[i, 'position'] -= sold
    candles.loc[i, 'action'] = -action 

    return candles, cash
```

```python
def execute_buy(candles, cash, i, to_buy):
    asks = candles.loc[i, 'asks']
    asks = np.array(ast.literal_eval(asks))
    j = 0
    bought = 0
    spent = 0
    
        
    while j < asks.shape[0] - 1 and bought + float(asks[j, 1]) < to_buy :
        bought += float(asks[j, 1])
        spent += float(asks[j, 0]) * float(asks[j, 1])
        j += 1


    candles.loc[i, 'impacted_price'] = float(asks[j, 0])



    cash -= spent 
    candles.loc[i, 'position'] += bought

    return candles, cash

def execute_sell(candles, cash, i, to_sell):
    bids = candles.loc[i, 'bids']
    bids = np.array(ast.literal_eval(bids))  
    j = 0
    sold = 0
    gained = 0

    while j < bids.shape[0] - 1 and sold + float(bids[j, 1]) < to_sell:
        sold += float(bids[j, 1])
        gained += float(bids[j, 0]) * float(bids[j, 1])
        j += 1


    candles.loc[i, 'impacted_price'] = float(bids[j, 0])

    
    cash += gained 
    candles.loc[i, 'position'] -= sold  

    return candles, cash
```

```python
candles['ideal_position'] = 0
candles['ideal_pnl'] = 0

def calculate_position_ideal(candles):


    cash = 0

    for i in tqdm.tqdm(range(len(candles))):


        if i >= 2:
            candles.loc[i, 'ideal_position'] = candles.loc[i-1, 'ideal_position']
            candles.loc[i, 'ideal_pnl'] = candles.loc[i, 'ideal_position'] * price + cash

        price = candles.loc[i, 'mid_price']
        position = candles.loc[i, 'ideal_position']

        if price < candles.loc[i,'buy_stop_loss']:
            if position > 0:
                cash += position * price
                candles.loc[i, 'ideal_position'] -= position
            continue     
        
        if price > candles.loc[i, 'buy_take_profit'] and position > 0:
            cash += position * price
            candles.loc[i, 'ideal_position'] -= position     
                                                    
        if price <  candles.loc[i, 'buy'] and position < 2.9 and candles.loc[i, 'buy'] > 0.01:
            cash -= (3 - position) * price
            candles.loc[i, 'ideal_position'] += 3 - position 
            position = 3           

                                
        if price > candles.loc[i,'sell_stop_loss']:
            if position < 0:
                cash -= abs(position * price)
                candles.loc[i, 'ideal_position'] = 0
            continue                                      
        
        if price < candles.loc[i, 'sell_take_profit'] and position < 0:
            cash -= abs(position * price)
            candles.loc[i, 'ideal_position'] = 0
                                     
        
        if price >  candles.loc[i, 'sell'] and position > -2.9 and candles.loc[i, 'sell'] > 0.01:
            cash += abs(-3 - position) * price
            candles.loc[i, 'ideal_position'] -= (3 - abs(position))
            position = -3                           


    return candles  
```

```python
def calculate_position(candles):

    bought, sold = 0, 0
    cash = 0

    for i in tqdm.tqdm(range(len(candles))):


        if i >= 2:
            candles.loc[i, 'position'] = candles.loc[i-1, 'position']
            candles.loc[i, 'PnL'] = candles.loc[i, 'position'] * price + cash

        price = candles.loc[i, 'mid_price']
        position = candles.loc[i, 'position']

        if price < candles.loc[i,'buy_stop_loss'] and position > 0:
            candles, cash =  execute_sell(candles, cash, i, to_sell = position)
            continue                                         
        
        if price > candles.loc[i, 'buy_take_profit'] and position > 0:
            candles, cash =  execute_sell(candles, cash, i, to_sell = position)
            #continue        
                                                    
        if price <  candles.loc[i, 'buy'] and position < 2.99 and candles.loc[i, 'buy'] > 0.01:
            candles, cash =  execute_buy(candles, cash, i, to_buy = 3 - position)
            #continue
                                
        if price > candles.loc[i,'sell_stop_loss'] and position < 0:
            candles, cash =  execute_buy(candles, cash, i, to_buy = abs(position))
            continue                                                            
        
        if price < candles.loc[i, 'sell_take_profit'] and position < 0:
            candles, cash =  execute_buy(candles, cash, i, to_buy = abs(position))
            #continue 
                                     
        
        if price >  candles.loc[i, 'sell'] and position > -2.99 and candles.loc[i, 'sell'] > 0.01:               
            candles, cash =  execute_sell(candles, cash, i, to_sell = 3 - abs(position))
            #continue

    return candles  
```

```python
def calculate_position_control(candles):

    bought, sold = 0, 0
    cash = 0

    for i in tqdm.tqdm(range(len(candles))):


        if i >= 2:
            candles.loc[i, 'position'] = candles.loc[i-1, 'position']
            candles.loc[i, 'PnL_control'] = candles.loc[i, 'position'] * price + cash

        price = candles.loc[i, 'mid_price']
        position = candles.loc[i, 'position']

        if price < candles.loc[i,'buy_stop_loss'] and position > 0:
            candles, cash =  execute_sell_control(candles, cash, i, to_sell = position)
            continue                                         
        
        if price > candles.loc[i, 'buy_take_profit'] and position > 0:
            candles, cash =  execute_sell_control(candles, cash, i, to_sell = position)
            #continue        
                                                    
        if price <  candles.loc[i, 'buy'] and position < 2.99 and candles.loc[i, 'buy'] > 0.01:
            candles, cash, action =  execute_buy_control(candles, cash, i, to_buy = 3 - position)
            #continue
                                
        if price > candles.loc[i,'sell_stop_loss'] and position < 0:
            candles, cash, action =  execute_buy_control(candles, cash, i, to_buy = abs(position))
            continue                                                            
        
        if price < candles.loc[i, 'sell_take_profit'] and position < 0:
            candles, cash, action =  execute_buy_control(candles, cash, i, to_buy = abs(position))
            #continue 
                                     
        
        if price >  candles.loc[i, 'sell'] and position > -2.99 and candles.loc[i, 'sell'] > 0.01:               
            candles, cash =  execute_sell_control(candles, cash, i, to_sell = 3 - abs(position))
            #continue

    return candles  
```

```python
ask_vol = lambda x: sum([float(y) for y in np.array(ast.literal_eval(x)) [:, 1]])
candles['volume'] = candles['bids'].apply(ask_vol) + candles['asks'].apply(ask_vol)
candles['traded'] = abs(candles['position'] - candles['position'].shift(-1))
candles['participation'] = candles['traded']/candles['volume']
```
