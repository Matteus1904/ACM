```python
def execute_buy(candles, cash, i, to_buy):
    bids = candles.loc[i, 'bids']
    bids = np.array(ast.literal_eval(bids))
    j = 0
    bought = 0
    spent = 0

    #for j in range(len(bids)):
    #    if bought + float(bids[j, 1]) < to_buy:
    #        break
    #    bought += float(bids[j, 1])
    #    spent += float(bids[j, 0]) * float(bids[j, 1])
        
    while bought + float(bids[j, 1]) < to_buy:
        bought += float(bids[j, 1])
        spent += float(bids[j, 0]) * float(bids[j, 1])
        #j += 1


    candles.loc[i, 'impacted_price'] = float(bids[j, 0])



    cash -= spent 
    candles.loc[i, 'position'] += bought

    return candles, cash

def execute_sell(candles, cash, i, to_sell):
    asks = candles.loc[i, 'asks']
    asks = np.array(ast.literal_eval(asks))  
    j = 0
    sold = 0
    gained = 0

#    for j in range(len(asks)):
##        if sold + float(asks[j, 1]) < to_sell:
#            break
#        sold += float(asks[j, 1])
#        gained += float(asks[j, 0]) * float(asks[j, 1])

    while sold + float(asks[j, 1]) < to_sell:
        sold += float(asks[j, 1])
        gained += float(asks[j, 0]) * float(asks[j, 1])
        #j += 1


    candles.loc[i, 'impacted_price'] = float(asks[j, 0])

    
    cash += gained 
    candles.loc[i, 'position'] -= sold  

    return candles, cash
```

```python
def calculate_position(candles):

    bought, sold = 0, 0
    cash = 0

    for i in tqdm.tqdm(range(len(candles))):


        if i >= 2:
            candles.loc[i, 'position'] = candles.loc[i-1, 'position']
            candles.loc[i, 'PnL'] = candles.loc[i, 'position'] * price + cash

        price = candles.loc[i, 'mid_price']
        position = candles.loc[i, 'position']

        if price < candles.loc[i,'buy_stop_loss'] and position > 0:
            candles, cash =  execute_sell(candles, cash, i, to_sell = position)
            continue                                         
        
        if price > candles.loc[i, 'buy_take_profit'] and position > 0:
            candles, cash =  execute_sell(candles, cash, i, to_sell = position)
            #continue        
                                                    
        if price <  candles.loc[i, 'buy'] and position < 3 and candles.loc[i, 'buy'] > 0.01:
            candles, cash =  execute_buy(candles, cash, i, to_buy = 3 - position)
            #continue
                                
        if price > candles.loc[i,'sell_stop_loss'] and position < 0:
            candles, cash =  execute_buy(candles, cash, i, to_buy = abs(position))
            continue                                                            
        
        if price < candles.loc[i, 'sell_take_profit'] and position < 0:
            candles, cash =  execute_buy(candles, cash, i, to_buy = abs(position))
            #continue 
                                     
        
        if price >  candles.loc[i, 'sell'] and position > -3 and candles.loc[i, 'sell'] > 0.01:               
            candles, cash =  execute_sell(candles, cash, i, to_sell = 3 - abs(position))
            #continue

    return candles  
```

```python
candles['ideal_position'] = 0
candles['ideal_pnl'] = 0

def calculate_position_ideal(candles):


    cash = 0

    for i in tqdm.tqdm(range(len(candles))):


        if i >= 2:
            candles.loc[i, 'ideal_position'] = candles.loc[i-1, 'ideal_position']
            candles.loc[i, 'ideal_pnl'] = candles.loc[i, 'ideal_position'] * price + cash

        price = candles.loc[i, 'mid_price']
        position = candles.loc[i, 'ideal_position']

        if price < candles.loc[i,'buy_stop_loss'] and position > 0:
           cash += position * price
           candles.loc[i, 'ideal_position'] -= position
           continue     
        
        if price > candles.loc[i, 'buy_take_profit'] and position > 0:
           cash += position * price
           candles.loc[i, 'ideal_position'] -= position     
                                                    
        if price <  candles.loc[i, 'buy'] and position < 2.9 and candles.loc[i, 'buy'] > 0.01:
           cash -= (3 - position) * price
           candles.loc[i, 'ideal_position'] += 3 - position 
           position = 3           

                                
        if price > candles.loc[i,'sell_stop_loss'] and position < 0:
            cash -= abs(position * price)
            candles.loc[i, 'ideal_position'] = 0
            continue                                      
        
        if price < candles.loc[i, 'sell_take_profit'] and position < 0:
            cash -= abs(position * price)
            candles.loc[i, 'ideal_position'] = 0
                                     
        
        if price >  candles.loc[i, 'sell'] and position > -2.9 and candles.loc[i, 'sell'] > 0.01:
            cash += abs(-3 - position) * price
            candles.loc[i, 'ideal_position'] -= (3 - abs(position))
            position = -3                           


    return candles  
```
